```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
df1 = pd.read_csv("Pomiary1.txt", delimiter=" ", decimal=",")
df1.head()
```

```python
temp_c = df1["Temperatura[C]"].to_numpy()
temp_c
```

```python
Iwy = df1["Iwy[mA]"].to_numpy()
Iwy
```

```python
df2 = pd.read_csv("Pomiary2.txt", delimiter=" ", decimal=",")
df2.head()
```

```python
R_obc = df2["Robc[Ohm]"].to_numpy()
R_obc
```

```python
I_wy_2 = df2["Iwy[mA]"].to_numpy()
I_wy_2
```

```python
plt.plot(temp_c, Iwy, 'r')
plt.scatter(temp_c, Iwy)
plt.grid(True)
plt.title("$I_{wy}(T)$")
plt.xlabel("$Temperatura[^{\circ}C]$")
plt.ylabel("$I_{wy}[mA]$")
plt.savefig("IwyT.png", dpi = 300)
plt.show()
```

```python
plt.plot(R_obc, I_wy_2, 'r')
plt.scatter(R_obc, I_wy_2)
plt.grid(True)
plt.title("$I_{wy}(R_{obc})$")
plt.xlabel("$R_{obc}[\Omega]$")
plt.ylabel("$I_{wy}[mA]$")
plt.savefig("IwyRobc.png", dpi = 300)
plt.show()
```
