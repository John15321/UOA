```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("KORONA.TXT", skiprows=20, header=None, delimiter="\t")
# df.columns = ["Value", "Unit", "Current", "Time", "Nwm co to"]
df
```

```python
df = df.drop(0, 1)
```

```python
# df.columns = ['Value', 'Unit', 'Current', 'Time', 'Nwm']
df.rename(columns={1:'Value', 2:'Unit', 3:'Current', 4:'Time', 5:'Nwm'}, inplace=True)
df
```

```python
mV = df["Value"].to_numpy()
```

```python
plt.plot(np.arange(0, len(mV)), mV)
plt.grid(True)
plt.xlabel("Próbki")
plt.ylabel("U[mV]")
plt.title("U(probek)")
plt.show()
```
